```python
from bs4 import BeautifulSoup
import requests #imports requests from other libraries
import time
import pandas as pd

pro_list=[]
comp_list=[]
loc_list=[]
```

```python
x = input("Enter job profile: ").replace(' ','+')
y = input("Enter location profile: ").replace(' ','+')
html_text = requests.get('https://www.careerbuilder.com/jobs?keywords='+x+'&location='+y).text


soup = BeautifulSoup(html_text, 'lxml')#lxml is the parser

list = soup.find("div", id = "jobs_collection")

jobs = list.find_all("div", class_ = "col big col-mobile-inline")


#print(job)
for job in jobs:
    profile = job.find("div", class_ = "data-results-title dark-blue-text b").text
    details = job.find("div", class_ = "data-details") #this part has three spans in a row, so i have to take 
    comp_and_loc = (details.find_all("span"))#          them all and parse them
    company = comp_and_loc[0].text
    location = comp_and_loc[1].text

    pro_list.append(profile)
    #print("Profile: " + profile)
    comp_list.append(company)
    #print("Company: " + company)
    loc_list.append(location)
    #print("Location: " + location)
```

```python
df = pd.DataFrame({'Profile': pro_list,
                   'Company': comp_list,
                   'Location': loc_list})
```

```python
big_companies = ['Apple', 'Microsoft', 'Google', 'Amazon', 'Facebook', 'Tesla']
```

```python
for index, row in df.iterrows():
    if row['Company'] in big_companies:
        df.drop(index)
```
